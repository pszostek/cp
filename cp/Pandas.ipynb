```python
res = merge(disass_df, bb_exec, left_on='bb', right_index=True, copy=False, how="outer")

res.set_index(['bb', 'inst_offset'])
for col in res.columns:
    if col != 'execcount' and col != 'noperands' and col != 'XED_ISA_SET' and col != "icount":
        res.drop(col, axis=1, inplace=True)
```

```python
grp = res.groupby(['XED_ISA_SET','noperands'])
```
